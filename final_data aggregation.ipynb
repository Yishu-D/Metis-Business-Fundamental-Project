```python
import pandas as pd
```

### Passer

```python
p_df = pd.read_csv('passer.csv')
```

```python
p_df.head()
```

```python
p_df['avgLength'] = p_df.groupby(['playerId'])['passLength'].transform('mean')
p_df.head()
```

```python
p_df['sum_passAtt'] = p_df.groupby(['playerId'])['passAtt'].transform('sum')
p_df['sum_passComp'] = p_df.groupby(['playerId'])['passComp'].transform('sum')
p_df['sum_passInt'] = p_df.groupby(['playerId'])['passInt'].transform('sum')
p_df.head()
```

```python
p_df.drop(columns=['passOutcomes','passLength','passAtt','passComp','passInt'], inplace = True)
p_df.head()
```

```python
p_df.drop_duplicates(inplace = True)
```

```python
p_df.head()
```

```python
p_df.to_csv('new_passer.csv', index = False, header=True)
```

### Receiver

```python
r_df = pd.read_csv('receiver.csv')
r_df.head()
```

```python
r_df['avgTotalYards'] = r_df.groupby(['playerId'])['recYards'].transform('mean')
r_df['avgRec'] = r_df.groupby(['playerId'])['rec'].transform('mean')
r_df['avgYac'] = r_df.groupby(['playerId'])['recYac'].transform('mean')
r_df.head()
```

```python
r_df.drop(columns = ['recYards','rec','recYac'], inplace=True)
r_df.info()
```

```python
r_df.head()
```

```python
r_df.drop_duplicates(inplace = True)
r_df.info()
```

```python
r_df.to_csv('new_receiver.csv', index = False, header=True)
```
